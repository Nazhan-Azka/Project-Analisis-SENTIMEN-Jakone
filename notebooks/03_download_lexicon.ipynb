# Download InSet Lexicon

Notebook ini digunakan untuk dokumentasi download otomatis file InSet Lexicon pada project skripsi **Analisis Sentimen pada Aplikasi JakOne Mobile Menggunakan Metode IndoBERT**.

Tahap ini hanya mengunduh dan mengecek file lexicon. Belum ada proses labeling sentimen, split dataset, training, evaluasi, atau visualisasi.

## File yang Diunduh

File InSet Lexicon diunduh dari raw URL GitHub berikut:

```text
https://raw.githubusercontent.com/fajri91/InSet/master/positive.tsv
https://raw.githubusercontent.com/fajri91/InSet/master/negative.tsv
```

File disimpan ke:

```text
data/lexicon/positive.tsv
data/lexicon/negative.tsv
```

## Opsi Overwrite

Pada script `src/03_download_lexicon.py`, variabel berikut mengatur apakah file lama akan ditimpa:

```python
OVERWRITE = False
```

Jika `False`, file yang sudah ada tidak akan diunduh ulang. Jika `True`, file lama akan ditimpa dengan hasil download terbaru.

````python
from pathlib import Path
import importlib.util
````

## Memuat Script Download

Notebook ini menggunakan fungsi dari `src/03_download_lexicon.py` agar hasilnya sama dengan menjalankan script lewat terminal.

````python
module_path = Path("src/03_download_lexicon.py")
if not module_path.exists():
    module_path = Path("../src/03_download_lexicon.py")

spec = importlib.util.spec_from_file_location("download_lexicon_module", module_path)
download_lexicon_module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(download_lexicon_module)
````

## Menjalankan Download Lexicon

````python
download_lexicon_module.main()
````

Jika download berhasil, output akan menampilkan path file, jumlah kata positif dan negatif, nama kolom, serta 5 baris pertama dari masing-masing file.
