## 1. Persiapan Path Project

Cell ini memastikan notebook dijalankan dari root project sehingga path dataset, model, dan folder output dapat ditemukan dengan konsisten.

```python
from pathlib import Path
import os

project_root = Path.cwd()
if project_root.name == "notebooks":
    project_root = project_root.parent
os.chdir(project_root)

project_root
```

## 2. Load Modul Evaluasi

Notebook memuat fungsi dari `src/08_evaluate_indobert.py` agar proses evaluasi di notebook dan script terminal memakai logika yang sama.

```python
import importlib.util

module_path = Path("src/08_evaluate_indobert.py")
spec = importlib.util.spec_from_file_location("evaluate_indobert", module_path)
evaluate_indobert = importlib.util.module_from_spec(spec)
spec.loader.exec_module(evaluate_indobert)

print("Dataset:", evaluate_indobert.INPUT_PATH)
print("Model:", evaluate_indobert.MODEL_DIR)
print("Label mapping:", evaluate_indobert.LABEL_MAPPING_PATH)
```

## 3. Jalankan Evaluasi Data Test

Cell ini membaca dataset final, mengambil hanya `split_set == "test"`, memuat tokenizer dan model dari folder lokal, melakukan batch inference, menghitung metrik, lalu menyimpan seluruh output evaluasi.

```python
metrics_summary = evaluate_indobert.evaluate_indobert()
metrics_summary
```

## 4. Ringkasan Metrik Utama

Cell ini menampilkan jumlah data test, accuracy, macro F1, dan weighted F1 dari file ringkasan evaluasi.

```python
print("Jumlah data test:", metrics_summary["test_count"])
print("Accuracy test:", round(metrics_summary["accuracy"], 6))
print("Macro F1 test:", round(metrics_summary["macro_f1"], 6))
print("Weighted F1 test:", round(metrics_summary["weighted_f1"], 6))
```

## 5. Classification Report

Cell ini membaca classification report versi CSV agar precision, recall, dan F1-score per kelas mudah diperiksa di notebook.

```python
import pandas as pd

report_df = pd.read_csv(evaluate_indobert.REPORT_CSV_PATH, index_col=0)
report_df
```

## 6. Confusion Matrix

Cell ini menampilkan confusion matrix dalam bentuk tabel untuk melihat pola prediksi benar dan salah pada setiap kelas sentimen.

```python
confusion_matrix_df = pd.read_csv(evaluate_indobert.CONFUSION_MATRIX_CSV_PATH, index_col=0)
confusion_matrix_df
```
