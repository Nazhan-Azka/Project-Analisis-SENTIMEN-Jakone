## 1. Persiapan Path Project

Cell ini memastikan notebook berjalan dari root project sehingga seluruh path input dan output dapat ditemukan secara konsisten.

```ipython
from pathlib import Path
import os

project_root = Path.cwd()
if project_root.name == "notebooks":
    project_root = project_root.parent
os.chdir(project_root)

project_root
```

## 2. Load Modul Visualisasi dan Analisis

Notebook memuat fungsi dari `src/09_visualization_analysis.py` agar hasil notebook sama dengan hasil script terminal.

```ipython
import importlib.util

module_path = Path("src/09_visualization_analysis.py")
spec = importlib.util.spec_from_file_location("visualization_analysis", module_path)
visualization_analysis = importlib.util.module_from_spec(spec)
spec.loader.exec_module(visualization_analysis)

print("Dataset:", visualization_analysis.DATASET_PATH)
print("Training log:", visualization_analysis.TRAINING_LOG_PATH)
print("Prediksi test:", visualization_analysis.TEST_PREDICTIONS_PATH)
```

## 3. Jalankan Proses Visualisasi dan Analisis

Cell ini membuat seluruh file visualisasi, ringkasan salah prediksi, analisis keyword, word cloud jika tersedia, dan ringkasan akhir untuk Bab 4.

```ipython
result = visualization_analysis.run_visualization_analysis()
result["summary"]
```

## 4. Distribusi Dataset

Bagian ini menampilkan distribusi label sentimen, distribusi tahun, distribusi rating, dan distribusi split dataset dari data final.

```ipython
from IPython.display import Image, display

for key in [
    "distribution_label_sentiment",
    "distribution_reviews_by_year",
    "distribution_rating",
    "distribution_split_set",
]:
    path = visualization_analysis.FIGURE_PATHS[key]
    if Path(path).exists():
        display(Image(filename=path))
```

## 5. Visualisasi Training

Bagian ini menampilkan kurva loss, validation F1 macro, dan validation accuracy berdasarkan `training_log.csv`.

```ipython
for key in [
    "training_loss_curve",
    "validation_f1_macro_curve",
    "validation_accuracy_curve",
]:
    path = visualization_analysis.FIGURE_PATHS[key]
    if Path(path).exists():
        display(Image(filename=path))
```

## 6. Visualisasi Evaluasi Test

Bagian ini menampilkan confusion matrix final serta perbandingan precision, recall, dan F1-score per kelas.

```ipython
for key in [
    "final_confusion_matrix",
    "classification_metrics_per_class",
]:
    path = visualization_analysis.FIGURE_PATHS[key]
    if Path(path).exists():
        display(Image(filename=path))
```

## 7. Analisis Salah Prediksi

Bagian ini menampilkan ringkasan pasangan actual-predicted yang salah serta contoh data salah prediksi.

```ipython
import pandas as pd

misclassification_summary = pd.read_csv(visualization_analysis.ANALYSIS_PATHS["misclassification_summary"])
misclassified_predictions = pd.read_csv(visualization_analysis.ANALYSIS_PATHS["misclassified_predictions"])

display(misclassification_summary)
display(misclassified_predictions.head(10))

path = visualization_analysis.FIGURE_PATHS["misclassification_by_pair"]
if Path(path).exists():
    display(Image(filename=path))
```

## 8. Analisis Keyword Isu Aplikasi

Bagian ini menampilkan keyword isu yang paling sering muncul pada seluruh data dan pada sentimen negatif.

```ipython
keyword_summary = pd.read_csv(visualization_analysis.ANALYSIS_PATHS["keyword_issue_summary"])
keyword_by_label = pd.read_csv(visualization_analysis.ANALYSIS_PATHS["keyword_issue_by_label"])

display(keyword_summary.head(10))
display(keyword_by_label[keyword_by_label["label"] == "negatif"].sort_values("count", ascending=False).head(10))

for key in ["top_keyword_issues", "top_negative_keyword_issues"]:
    path = visualization_analysis.FIGURE_PATHS[key]
    if Path(path).exists():
        display(Image(filename=path))
```
