# Persiapan Tahap 3 - Pengecekan InSet Lexicon

Notebook ini digunakan untuk dokumentasi persiapan file InSet Lexicon pada project skripsi **Analisis Sentimen pada Aplikasi JakOne Mobile Menggunakan Metode IndoBERT**.

Tahap ini hanya mengecek ketersediaan dan format file lexicon. Belum ada proses labeling sentimen, split dataset, training, evaluasi, atau visualisasi.

## Lokasi File Lexicon

File InSet Lexicon perlu diletakkan secara manual pada folder berikut:

```text
data/lexicon/
```

File yang diperlukan:

```text
data/lexicon/positive.tsv
data/lexicon/negative.tsv
```

Isi file `positive.tsv` dan `negative.tsv` tidak dibuat manual di project ini. File tersebut diunduh sendiri dari sumber InSet Lexicon.

## Tujuan Pengecekan

Pengecekan dilakukan untuk memastikan:

1. Folder `data/lexicon/` tersedia.
2. File `positive.tsv` dan `negative.tsv` tersedia.
3. Kedua file dapat dibaca sebagai TSV dengan separator tab.
4. Jumlah kata, nama kolom, dan contoh isi file dapat ditampilkan.
5. Kolom kata dan bobot dapat terdeteksi sebelum tahap labeling dilakukan.

````python
from pathlib import Path
import importlib.util
````

## Memuat Script Pengecekan

Notebook ini menggunakan fungsi dari `src/03_check_lexicon.py` agar hasil pengecekan sama dengan script terminal.

````python
module_path = Path("src/03_check_lexicon.py")
if not module_path.exists():
    module_path = Path("../src/03_check_lexicon.py")

spec = importlib.util.spec_from_file_location("lexicon_check_module", module_path)
lexicon_check_module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(lexicon_check_module)
````

## Menjalankan Pengecekan Lexicon

````python
lexicon_check_module.main()
````

Jika file lexicon sudah benar, output akan menampilkan jumlah kata pada `positive.tsv` dan `negative.tsv`, nama kolom, status format, serta 5 baris pertama dari masing-masing file.
